```python
import pandas as pd
import numpy as np
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['figure.dpi'] = 300
plt.rcParams['figure.figsize'] = [12,4.5]
```

```python
fact = lambda x: np.math.factorial(x)
```

```python
# Equation
N = 16
p = q = 1/2
n = list(range(N+1))
m = [2*n-N for n in n]
PN = [((0.5**N)*fact(N))/(fact(0.5*(N+m))*fact(0.5*(N-m))) for m in m]
```

```python
# Plotting
sns.barplot(x = m, y = PN)

plt.title("Binomial Distribution",fontsize = 17)
plt.xlabel("m")
plt.ylabel("Probability")

plt.tight_layout()
```
