```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

from scipy.integrate import odeint

plt.style.use('seaborn-notebook')
plt.rcParams['figure.dpi'] = 100
np.set_printoptions(suppress=True)
```

Sky diving problem:
$\frac{d \vec S}{dt} = \vec v = v_x\hat x + v_y \hat y $

```python
def dvdt(V, t):
    
    vx,vy = V[0],V[1]
    
    dvx = -b/m * np.sqrt(vx**2 + vy**2) * vx         #vx'(t)
    dvy = -g - (b/m) * np.sqrt(vx**2 + vy**2) * vy   #vy'(t)
    
    return [dvx, dvy]
```

```python
v0 = [50, 0]
m  = 60
vt = -55
g  = 9.81
b  = m * g / vt**2 
t  = np.linspace(0, 20, 100)
```

```python
y = odeint(dsdt,v0,t)
```

```python
vx = y.T[0]
vy = y.T[1]
```

```python
plt.plot(t,vx);
plt.xlabel('$t$')
plt.ylabel('$v_x(t)$');
```

```python
plt.plot(t,vy)
plt.xlabel('$t$')
plt.ylabel('$v_y(t)$')
plt.axhline(vt, color = 'r', ls = '--');
```
